# Topic 03 - Problem 3: Calculate Standard Deviation and Variance

---

## 1. About the Problem

This problem asks me to calculate the **standard deviation** and **variance** for a given numerical dataset.  
The **variance** measures how much the values in a dataset deviate from the mean, while the **standard deviation** is the square root of the variance.  
Both are used to understand the **spread** of the data â€” a higher value means more spread, and a lower value means the data points are close to the mean.  
To solve this, I will calculate the **mean**, then compute the **variance** and **standard deviation** using standard formulas.

---



## 2. Solution Code

```python
import math
def calculate_variance_and_std(data):
    if not data:
        return "dataset is empty"
    
    mean=sum(data)/len(data)
    squared_diff=[(x-mean)**2 for x in data]
    variance=sum(squared_diff)/len(data)
    std=math.sqrt(variance)

    return{
        'variance':variance,
        'standard_deviation':std
    }
data = [45, 50, 38, 60, 70, 55, 65]

stats = calculate_variance_and_std(data)
print("Variance and Standard Deviation:", stats)
```
