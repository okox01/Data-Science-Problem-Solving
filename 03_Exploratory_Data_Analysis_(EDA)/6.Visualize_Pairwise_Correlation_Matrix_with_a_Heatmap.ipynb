## 1. About the Problem

This problem asks me to visualize the pairwise correlation matrix using a **heatmap**.  
A **heatmap** is an excellent way to represent a matrix of values, and in this case, it helps show the strength and direction of relationships between different variables.  
By using color gradients, I can easily identify strong, weak, or no correlations.  
To solve this, I will use the `seaborn` library to generate the heatmap and display the correlation matrix I computed earlier.

---


## 2. Solution Code

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
def plot_correlation_heatmap(data):
    df=pd.DataFrame(data)

    matrix_plot=df.corr()

    plt.figure(figsize=(8,6))
    sns.heatmap(matrix_plot,annot=True,cmap='coolwarm',fmt='.3f',linewidths=0.5)
    plt.title("pairwise correlation")
data = {
    "age": [25, 30, 28, 60, 70, 55, 65],
    "salary": [50000, 60000, 55000, 68000, 75000, 57000, 63000],
    "experience": [2, 5, 3, 10, 15, 8, 12]
}

plot_correlation_heatmap(data)
```
