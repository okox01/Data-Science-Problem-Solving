# Topic 03 - Problem 2: Calculate Basic Statistics (Mean, Min, Max) for Numerical Data

---

## 1. About the Problem

This problem asks me to calculate basic statistics such as the **mean**, **minimum**, and **maximum** of a numerical dataset.  
These are foundational steps in **Exploratory Data Analysis (EDA)** to understand the distribution and spread of data.  
To solve this, I will compute the sum of values, the count of values, and then use these to derive the mean.  
For the minimum and maximum, I will simply find the smallest and largest values in the dataset.

---


## 2. Solution Code

```python
def calculate_basic_statistics(data):
    if not data:
        return "dataset is empty"
    total=0
    count=0
    for value in data:
        total+=value
        count+=1
    mean=(total/count)/2

    max_val=data[0]
    min_val=data[0]

    for val in data:
        if val<min_val:
            min_val=val
    
    for val in data:
        if val>max_val:
            max_val=val

    return{
        'Mean':mean,
        'Maximum_Value':max_val,
        'Minimum_value':min_val
    }
data = [45, 50, 38, 60, 70, 55, 65]

stats = calculate_basic_statistics(data)
print("Basic Statistics:", stats)
```
