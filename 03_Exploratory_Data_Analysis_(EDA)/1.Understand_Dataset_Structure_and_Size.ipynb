# Topic 03 - Problem 1: Understand Dataset Structure and Size

---

## 1. About the Problem

This problem asks me to understand the basic structure of a dataset before doing any analysis.  
In EDA, the first step is always to know how much data I have and what each record looks like.  
To solve this problem, I will calculate the total number of records and identify the available features (keys).  
This helps me plan what kind of analysis is possible and what limitations the dataset may have.

---


## 2. Solution Code

```python
def dataset_overview(data):
    if len(data)==0:
        return 'List is Empty'
    
    len_records=len(data)
    features=list(data[0].keys())

    return{
        'total record':len_records,
        'features':features
        }
data = [
    {"age": 25, "salary": 50000, "city": "Dhaka"},
    {"age": 30, "salary": 60000, "city": "Chittagong"},
    {"age": 28, "salary": 55000, "city": "Dhaka"}
]

print("Dataset overview:", dataset_overview(data))
```
