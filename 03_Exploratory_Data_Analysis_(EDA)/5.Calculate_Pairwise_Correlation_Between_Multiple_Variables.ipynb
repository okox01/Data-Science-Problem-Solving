# Topic 03 - Problem 5: Calculate Pairwise Correlation Between Multiple Variables

---

## 1. About the Problem

This problem asks me to calculate the **pairwise correlation** between multiple variables in a dataset.  
In real-world data, it's crucial to understand how each feature relates to the others.  
Pairwise correlation is often visualized using a **correlation matrix**, where each cell represents the correlation between two variables.  
To solve this, I will calculate the correlation between all pairs of numerical variables in a dataset.

---


## 2. Solution Code

```python
import pandas as pd
import seaborn as sns
def calculate_pairwise_correlation(data):
    df = pd.DataFrame(data)
    return sns.heatmap(df.corr(),annot=True)

data = {
    "age": [25, 30, 28, 60, 70, 55, 65],
    "salary": [50000, 60000, 55000, 68000, 75000, 57000, 63000],
    "experience": [2, 5, 3, 10, 15, 8, 12]
}

correlation_matrix = calculate_pairwise_correlation(data)
print("Pairwise Correlation Matrix:")
print(correlation_matrix)
```
