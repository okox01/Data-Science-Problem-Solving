# Problem 1: Calculate Mean Temperature

---

## 1. About the Problem

This problem asks me to find the average temperature from a list of daily temperatures.  
To solve this, I need to add all the temperature values together and then divide the sum by the total number of temperatures.  
I will also round the result to one decimal place to make it easier to read.  
This is important because calculating averages is a basic and common task in data science when summarizing data.

---


## 2. Solution Code

```python
def average(temps):
    total=0
    s=0
    for i in temps:
        s+=i
        total+=1
    mean = s / total
    return round(mean,1)

temps = [30, 32.56, 31, 29, 35, 33, 34]
print("Average Temperature:",average(temps))
```
