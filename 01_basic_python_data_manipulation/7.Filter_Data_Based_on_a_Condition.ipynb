# Problem 7: Filter Data Based on a Condition

---

## 1. About the Problem

This problem asks me to filter values from a list based on a specific condition.  
Filtering data is a very common task in data science, especially during data cleaning and preprocessing.  
To solve this problem, I will iterate through the list and select only the values that satisfy the given condition.  
This helps remove irrelevant or noisy data before analysis or model training.

---


## 2. Solution Code

```python
def filter_greater_than_threshold(values,threshold):
    filtered_values=[]
    for data in values:
        if data>threshold:
            filtered_values.append(data)

    return filtered_values

values = [5, 12, 7, 20, 3, 15, 8]
threshold=10
print("values greater than threshold:",filter_greater_than_threshold(values,threshold))
```
