# Problem 14: Min-Max Normalization of Data

---

## 1. About the Problem

This problem asks me to normalize a numerical dataset using **Min-Max scaling**.  
Normalization is necessary because machine learning algorithms perform better when features are on a similar scale.  
In Min-Max normalization, each value is transformed to a range between 0 and 1 using the formula:

(value − min) / (max − min)

This technique is widely used in data preprocessing before training ML models.

---


## 2. Solution Code

```python
def normalizes(data):
    min=data[0]
    max=data[0]

    for v in data:
        if v<min:
            min=v
    for x in data:
        if x>max:
            max=x
    normalized_data=[]
    for value in data:
        norm=(value-min)/(max-min)
        normalized_data.append(norm)
    
    return normalized_data
values = [50, 60, 70, 80, 90, 100]
print("Normalized values: ",normalizes(values))
```
