# Problem 2: Find the Most Frequent Element in a List

---

## 1. About the Problem

This problem asks me to find the element that appears most frequently in a list of values.  
To solve this, I will count how many times each element appears using a dictionary, then find the element with the highest count.  
Finding the most frequent value, or mode, is important in data science because it helps summarize categorical data and identify common patterns.

---


## 2. Solution Code

```python
def most_frequent_element(lst):
    frequency={}
    for item in lst:
        frequency[item]=frequency.get(item,0)+1
    most_frequent=max(frequency.values())
    for key, value in frequency.items():
        if value==most_frequent:
            return key
# Testing the function
data = [3, 5, 3, 2, 9, 5, 3, 5, 5, 2,2,2,2,2]
print("Most frequent element:", most_frequent_element(data))
```
