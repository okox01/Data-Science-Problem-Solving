# Problem 13: Detect Outliers Using Simple Threshold Logic

---

## 1. About the Problem

This problem asks me to identify outliers in a numerical dataset.  
Outliers are values that are unusually high or low compared to the rest of the data and can negatively affect analysis and machine learning models.  
To solve this problem, I will calculate the mean of the data and then flag values that differ from the mean by more than a fixed threshold.  
This is a simple but effective approach to understand outlier detection before moving to advanced methods like IQR or Z-score.

---


## 2. Solution Code

```python
def detect_outlier(data,threshold):
    total=0
    count=0
    for value in data:
        total+=value
        count+=1
    mean=total/count
    outlier=[]
    for x in data:
        if abs(x-mean)>threshold:
            outlier.append(x)
    return outlier

numbers = [10, 12, 11, 13, 100, 14, 12, 11]
threshold = 20
print("Outlier: ",detect_outlier(numbers,threshold))
```
