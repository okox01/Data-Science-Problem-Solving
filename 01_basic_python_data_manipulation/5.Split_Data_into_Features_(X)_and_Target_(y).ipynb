# Problem 5: Split Data into Features (X) and Target (y)

---

## 1. About the Problem

This problem asks me to separate a dataset into features (X) and a target variable (y).  
In data science and machine learning, models are trained using features as input and a target as output.  
To solve this problem, I will take a list of records where the last value represents the target, and the remaining values represent features.  
This is a fundamental step before training any machine learning model.

---



## 2. Solution Code

```python
def split_features_target(data):
    x=[]
    y=[]
    for row in data:
        x.append(row[:-1]) 
        y.append(row[-1])
    return x,y

dataset = [
    [1500, 3, 200000],
    [1800, 4, 250000],
    [1200, 2, 150000],
    [2000, 5, 300000]
]

X, y = split_features_target(dataset)
print("Features (X):", X)
print("Target (y):", y)
```
