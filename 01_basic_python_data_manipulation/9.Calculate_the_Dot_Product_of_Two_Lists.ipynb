# Problem 9: Calculate the Dot Product of Two Lists

---

## 1. About the Problem

This problem asks me to calculate the dot product of two lists (vectors).  
The dot product is an important mathematical operation used in data science and machine learning, especially in linear regression, similarity measures, and neural networks.  
To solve this problem, I will multiply corresponding elements from both lists and then sum all the results.  
This helps me understand how vector-based calculations work at a low level.

---


## 2. Solution Code

```python
def dot_product(l1,l2):
    if len(l1) != len(l2):
        raise ValueError("Vectors must be of same length")
    
    result=0
    for x,y in zip(l1,l2):
        result+=x*y
    return result

vector_a=[1,2,3]
vector_b=[4,5,6]
print("Dot Product:",dot_product(vector_a,vector_b))
```
