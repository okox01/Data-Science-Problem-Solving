# Problem 10: Standardize a List of Numbers (Z-score Normalization)

---

## 1. About the Problem

This problem asks me to standardize a list of numerical values using Z-score normalization.  
Standardization is very important in machine learning because many algorithms assume data is centered around zero with equal variance.  
To solve this problem, I will calculate the mean and standard deviation of the list, then transform each value using the Z-score formula.  
This helps make features comparable and improves model performance.

---


## 2. Solution Code

```python
import math
def z_score_standardization(values):
    mean=sum(values)/len(values)
    standardized_data=[]
    variance=sum((x-mean)**2 for x in values)/len(values)
    std=math.sqrt(variance)
    for z in values:
        a=(z-mean)/std
        standardized_data.append(round(a,3))
    
    return standardized_data

# Testing the function
values = [10, 20, 30, 40, 50]
print("Standardized values:", z_score_standardization(values))
```
