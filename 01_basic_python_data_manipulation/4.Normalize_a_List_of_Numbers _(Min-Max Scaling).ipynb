# Problem 4: Normalize a List of Numbers (Min-Max Scaling)

---

## 1. About the Problem

This problem asks me to normalize a list of numbers using Min-Max scaling.  
Normalization is important in data science and machine learning because many algorithms perform better when features are on the same scale.  
To solve this problem, I will find the minimum and maximum values in the list and then apply the Min-Max formula to each element.  
This transforms all values to a range between 0 and 1.

---


## 2. Solution Code

```python
def min_max_normalization(values):
    max_value=max(values)
    min_value=min(values)
    normalized=[]
    for item in values:
        norm_value=(item-min_value)/(max_value-min_value)
        normalized.append(round(norm_value,3))
    return normalized
    
values = [10, 20, 30, 40, 50]
print("Normalized values:", min_max_normalization(values))
```
