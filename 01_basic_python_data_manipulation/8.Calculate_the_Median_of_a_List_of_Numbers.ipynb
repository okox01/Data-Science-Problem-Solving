# Problem 8: Calculate the Median of a List of Numbers

---

## 1. About the Problem

This problem asks me to calculate the median of a list of numbers.  
The median is an important statistical measure in data science because it represents the middle value of a dataset and is less affected by outliers than the mean.  
To solve this problem, I will first sort the list and then find the middle value.  
If the list has an even number of elements, I will calculate the average of the two middle values.

---


## 2. Solution Code

```python
def median(data):
    sorted_data=sorted(data)
    mid=len(sorted_data)//2

    if len(sorted_data)%2==0:
        return (sorted_data[mid-1]+sorted_data[mid])/2
    else:
        return sorted_data[mid]


values_odd = [7, 3, 5, 1, 9]
values_even = [10, 2, 4, 6]
print("Median of odd number of data points:",median(values_odd))
print("Median of even number of data points:",median(values_even))
```
