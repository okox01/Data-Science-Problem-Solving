# Problem 6: Count Occurrences of Each Element in a List

---

## 1. About the Problem

This problem asks me to count how many times each element appears in a list.  
In data science, frequency counting is very important for understanding distributions, categorical data, and class balance.  
To solve this problem, I will iterate through the list and store counts using a dictionary, where keys represent elements and values represent their frequencies.  
This approach helps summarize data efficiently.

---


## 2. Solution Code

```python
def occurances(data):
    dict={}
    for item in data:
        dict[item]=dict.get(item,0)+1
    
    return dict
data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
print("Occurances of the words:",occurances(data))
```
