# Problem 12: Handle Missing Values by Removing Incomplete Records

---

## 1. About the Problem

This problem asks me to handle missing values in a dataset by removing records that contain missing data.  
In real-world datasets, missing values are very common and can cause errors or incorrect results if not handled properly.  
To solve this problem, I will iterate through the dataset and keep only those records that do not contain missing values (`None`).  
This approach is useful when the dataset is large enough and removing some rows does not significantly affect the analysis.

---


## 2. Solution Code

```python
def remove_none(data):
    cleaned_data=[]
    for record in data:
        if None not in record:
            cleaned_data.append(record)
    
    return cleaned_data
dataset = [
    [25, 50000],
    [30, None],
    [22, 42000],
    [28, 48000],
    [None, 39000]
]

print("Cleaned dataset: ",remove_none(dataset))
```
