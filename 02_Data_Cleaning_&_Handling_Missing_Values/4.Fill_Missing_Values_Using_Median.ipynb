# Topic 02 - Problem 4: Fill Missing Values Using Median

---

## 1. About the Problem

This problem asks me to replace missing numerical values using the median instead of the mean.  
Median is often a better choice when the data contains outliers because it is not affected by extreme values.  
To solve this problem, I will first sort the valid values, calculate the median, and then fill the missing values with it.  
This approach helps make the data more robust for analysis and modeling.

---


## 2. Solution Code

```python
def fill_missing_with_median(data):
    valid_data=[]
    for v in data:
        if v is not None:
            valid_data.append(v)
    
    valid_data3=sorted(valid_data)  # or; valid_data.sort()
    mid=len(valid_data3)//2
    if len(valid_data3)%2==0:
        median=(valid_data3[mid-1]+valid_data3[mid])/2
    else:
        median=valid_data3[mid]
    
    filled_data=[]
    for val in data:
        if val is None: 
            filled_data.append(median)
        else:
            filled_data.append(val)
    
    return filled_data

data = [100, None, 50, 200, None, 60]
print("Data after filling missing values: ",fill_missing_with_median(data))
```
