# Topic 02 - Problem 1: Identify Missing Values in a Dataset

---

## 1. About the Problem

This problem asks me to identify which records in a dataset contain missing values (`None`).  
Detecting missing data is the first step in data cleaning because missing values can affect analysis and model accuracy.  
To solve this, I will iterate through each record and check if it contains any missing value.  
This helps me know which parts of the data need attention or cleaning.

---


## 2. Solution Code

```python
def find_missing_records(data):
    missing_indices=[]
    for idx,record in enumerate(data):
        if None in record:
            missing_indices.append(idx)
    
    return missing_indices

dataset = [
    [25, 50000],
    [30, None],
    [22, 42000],
    [28, 48000],
    [None, 39000]
]
print("Missing Indices: ",find_missing_records(dataset))
```
