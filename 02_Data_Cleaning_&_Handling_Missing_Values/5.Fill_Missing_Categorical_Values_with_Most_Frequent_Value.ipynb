# Topic 02 - Problem 5: Fill Missing Categorical Values with Most Frequent Value

---

## 1. About the Problem

This problem asks me to handle missing categorical values (like gender, city, or category names).  
Numerical techniques like mean or median do not work for categorical data.  
A common and practical approach is to replace missing values with the most frequent (mode) value.  
To solve this problem, I will count the frequency of each category and then fill missing values using the most common one.

---


## 2. Solution Code

```python
def fill_missing_with_mode(data):
    frequency={}
    for v in data:
        if v is not None:
            frequency[v]=frequency.get(v,0)+1
    
    max_mode=max(frequency,key=frequency.get)
    
    filled_data=[]

    for val in data:
        if val is None:
            filled_data.append(max_mode)
        else:
            filled_data.append(val)

    return filled_data

categories = ["A", None, "B", "A", None, "C", "A"]
print("After filling missing data: ",fill_missing_with_mode(categories))
```
