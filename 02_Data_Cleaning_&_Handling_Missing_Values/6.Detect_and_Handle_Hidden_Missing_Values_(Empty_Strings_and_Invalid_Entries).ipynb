# Topic 02 - Problem 6: Detect and Handle Hidden Missing Values (Empty Strings and Invalid Entries)

---

## 1. About the Problem

This problem asks me to identify missing values that aren’t just `None` but may be hidden as empty strings (`""`), spaces, or invalid placeholders like `"NA"` or `"?"`.  
Real datasets often have these hidden missing values that can cause issues if not detected.  
To solve this, I will scan the data and treat these special cases as missing values, then replace them with a chosen fill value.

---


## 2. Solution Code

```python
def fill_hidden_values(data,fill_value):
    hidden_examples=['NA',""," ",None,"?"]
    cleaned_data=[]

    for v in data:
        if v in hidden_examples:
            cleaned_data.append(fill_value)
        else:
            cleaned_data.append(v)

    return cleaned_data 
data = ["A", "", "B", "NA", None, "C", "?", " "]
fill_with = "Unknown"

print("Cleaned Data: ",fill_hidden_values(data,fill_with))
```
