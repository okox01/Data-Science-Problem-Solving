# Topic 02 - Problem 3: Fill Missing Numerical Values with Mean

---

## 1. About the Problem

This problem asks me to handle missing numerical values by replacing them with the mean of the available data.  
Instead of removing records, filling missing values helps preserve important data.  
Mean imputation is commonly used when data is roughly evenly distributed.  
To solve this problem, I will first calculate the mean of non-missing values and then replace missing entries with that mean.

---


## 2. Solution Code

```python
def fill_missing_with_mean(data):
    total=0
    count=0

    for v in data:
        if v is not None:
            total+=v
            count+=1
    
    mean=total/count

    filled_data=[]

    for val in data:
        if val is None:
            filled_data.append(mean)
        else:
            filled_data.append(val)
    

    return filled_data

data = [50, None, 60, 55, None, 65]
print("Data after filling missing values: ",fill_missing_with_mean(data))
```
