# Topic 02 - Problem 2: Remove Records with Missing Values

---

## 1. About the Problem

This problem asks me to remove records that contain missing values (`None`) from a dataset.  
In many real-world datasets, some rows are incomplete and cannot be used for analysis or model training.  
To solve this problem, I will loop through each record and keep only those records that are fully complete.  
This is useful when missing data is small and removing rows does not significantly reduce the dataset size.

---


## 2. Solution Code


```python
def removing_missing_records(data):
    cleaned_data=[]
    for v in data:
        if None not in v:
            cleaned_data.append(v)

    return cleaned_data

dataset = [
    [25, 50000],
    [30, None],
    [22, 42000],
    [28, 48000],
    [None, 39000]
]

print("Cleaned dataset:", remove_missing_records(dataset))
```
